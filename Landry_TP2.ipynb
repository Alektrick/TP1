# SVM

```python
def make_meshgrid(x, y, h=.02):
    """Create a mesh of points to plot in

    Parameters
    ----------
    x: data to base x-axis meshgrid on
    y: data to base y-axis meshgrid on
    h: stepsize for meshgrid, optional

    Returns
    -------
    xx, yy : ndarray
    """
    x_min, x_max = x.min() - 1, x.max() + 1
    y_min, y_max = y.min() - 1, y.max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    return xx, yy

def plot_contours(ax, clf, xx, yy, **params):
    """Plot the decision boundaries for a classifier.

    Parameters
    ----------
    ax: matplotlib axes object
    clf: a classifier
    xx: meshgrid ndarray
    yy: meshgrid ndarray
    params: dictionary of params to pass to contourf, optional
    """
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    out = ax.contourf(xx, yy, Z, **params)
    return out
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import train_test_split
from sklearn import svm, datasets
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score
from sklearn.metrics import confusion_matrix
```

```python

def entrainerSVM(kernel, X_train, y_train, nbComposantes = 2):

  cv = 10
  kf = KFold(n_splits=cv)

  C = 1.0  # SVM regularization parameter
  model = svm.SVC(kernel=kernel, C=C)


  for fold, (train_index, valid_index) in enumerate(kf.split(X_train, y_train)):

    X_train_cv, X_valid_cv = X_train.iloc[train_index], X_train.iloc[valid_index]
    y_train_cv, y_valid_cv = y_train[train_index], y_train[valid_index]

    model.fit(X_train_cv, y_train_cv)
    y_pred_cv = model.predict(X_valid_cv)

    print("Fold ", fold, ", ", "Accuracy = ", accuracy_score(y_valid_cv, y_pred_cv))

  # Illustrer les marges si le nombre de composantes est egal à 2
  if(nbComposantes == 2):

    X0, X1 = X_train.iloc[:, 0], X_train.iloc[:, 1]
    xx, yy = make_meshgrid(X0, X1)

    fig, ax = plt.subplots()
    plot_contours(ax, model, xx, yy,
                      cmap=plt.cm.coolwarm, alpha=0.8)
    ax.scatter(X_train.iloc[:,0], X_train.iloc[:,1], c=y_train, cmap=plt.cm.coolwarm, s=20, edgecolors='k')
    #ax.set_xlim(xx.min(), xx.max())
    #ax.set_ylim(yy.min(), yy.max())
    #ax.set_xlabel('Sepal length')
    #ax.set_ylabel('Sepal width')
    #ax.set_xticks(())
    #ax.set_yticks(())
    #ax.set_title(title)

  return model
```

```python
def predireSVM(model, X_test, y_test):

  y_pred = model.predict(X_test)

  # Afficher les metriques

  metrique_accuracy = accuracy_score(y_test, y_pred)
  print("Accuracy : " + str(metrique_accuracy))
  #metrique_recall = recall_score(y_test, y_pred, average='weighted')
  #print("Recall : " + str(metrique_recall))
  #metrique_precision = precision_score(y_test, y_pred, average='weighted')
  #print("Precision : " + str(metrique_precision))
  metrique_f1score = f1_score(y_test, y_pred, average='weighted')
  print("F1 score : " + str(metrique_f1score))

  # Matrice de confusion

  print ("Catégories : ")
  print(np.unique(y_test))

  print ("Matrice de confusion : ")
  print(confusion_matrix(y_test, y_pred))
```

# Produit R1

```python
X_train, X_test, y_train, y_test = train_test_split(pca_prod1_comp2_df, y_prod1, test_size=0.2, random_state=42)

model_linear_prod1_comp2 = entrainerSVM('linear', X_train, y_train)
```
